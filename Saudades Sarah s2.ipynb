```
#importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```
df = pd.read_excel("https://github.com/DAVIDKSOUSA/DATA-SCIENCE-NA-PR-TICA/blob/master/SAUDADES%20SARAH.csv?raw=true")
df.head()
```

```
#eliminar as colunas com valores ausentes
words = df.dropna(subset=["words"], axis=0)["words"]

#concatenar as palavras
all_summary = " ".join(s for s in words)

#ver quatidade de palavras
print("Quantidade de palavras: {}".format(len(all_summary)))
```

```
# gerar uma wordcloud
wordcloud = WordCloud(background_color="black",
                      width=1600, height=800).generate(all_summary)
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);
wordcloud.to_file("airbnb_summary_wordcloud.png")
 
```

```


# endereo LOCAL da SUA imagem
rio_mask = np.array(Image.open("/content/flat-black-heart-icon-isolated-on-white-background-vector-illustration.jpg"))
   
# gerar uma wordcloud
wordcloud = WordCloud(
                      background_color="white",
                      width=1600, height=800, max_words=50000,
                      mask=rio_mask, max_font_size=50,
                      min_font_size=2).generate(all_summary)
 
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(15,15))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud)
wordcloud.to_file("airbnb_summary_wordcloud.png")
```
