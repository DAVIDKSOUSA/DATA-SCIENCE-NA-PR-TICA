###Método LOC

```
#séries são obejetos mais limitados do que DataFrames, os quais são maiores e mais completos.
#pressionando TAB irá demostrar quais métodos estão diaponíveis para utilização com séries ou DataFrames.
```

```
#consultar dados - localizar os dados, cujo índice é 1.
df.loc[1]
```

```
#identificando o tipo - neste caso é uma série
type(df.loc[1])
```

```
#utilizando o slice - irá identificar o intervalo.
df.loc[10:20]
```

```
#selecionar do passageiro 10 até o 20, pulando de 2 em 2
df.loc[10:20:2]
```

```
#visualizar as linhas 1 e 2 e apenas algumas colunas     
df.loc[[1,2], ["Name", "Sex", "Age" ]]
```

```
#todas as linhas e apenas algumas colunas
df.loc[:, ["Name", "Sex", "Age"]]
```

###Método Query!

```
#simplifica a utilização do método LOC
df.query("Age>20")
```

```
#operador e
df.query("Age>20 & Sex=='male'").head(10)
```

```
#operador or
df.query("Age>20 | Sex=='male'").head(10)
```

```
#operador in e o parâmetro inplace (altera o DataFrame em memória) - Pessoas que embarcaram nos postos B e C
#DataFrame em memória foi alterado
df.query("Embarked in ['C', 'Q']", inplace=True)
```
